# Train Model
This notebook details the model training process for the Removal Of BOgus Transients (ROBOT) pipeline, specifically the Real/Bogus Convolutional Neural Network.

## Imports

```python
import numpy as np
import matplotlib
import itertools
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import json

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc, confusion_matrix

from keras.models import Sequential, Model
from keras.layers import Activation, Dense, Dropout, Flatten, Input, Concatenate, Conv2D, MaxPooling2D, GlobalAveragePooling2D, AveragePooling2D
import keras.backend as K
from keras.callbacks import EarlyStopping

print("\n== Versions ==")
print("Tensorflow:", tf.__version__)
print("Keras:", keras.__version__)
```

## Import dataset
*dataset.npy* is a numpy file that contains the data and labels for training and testing. These data can be separated and split into training and testing sets.

Note: The data in *dataset.npy* has already been preprocessed and augmented. To see these steps, refer to the *Preprocessing* notebook.

```python
random_state = 42  # Random seed for reproducable results
dataset = np.load("./Data/dataset.npy", allow_pickle=True)  # Read in the dataset
data, labels = zip(*list(dataset))  # Separate image data from labels
data, labels = np.array(data), np.array(labels)  # Cast image data and labels to numpy arrays
X_train, X_test, y_train, y_test = train_test_split(
    data, labels, test_size=0.1, random_state=random_state)  # Split image data (X) and labels (y) into training & testing sets

print(f'Sample shape: {X_train[0].shape}')
print(f'Training: {len(X_train)} samples ({sum(y_train)} "Real" | {len(X_train)-sum(y_train)} "Bogus") \n Testing: {len(X_test)} samples ({sum(y_test)} "Real" | {len(X_test)-sum(y_test)} "Bogus")')
```

## Remove "blank" samples from existing data
Some images within the dataset consist solely of '0' value pixels. These samples are removed from the training & testing process.

Note: A total of 361 samples are removed in this dataset, which includes many augmented copies of blanks from the original dataset. In the original set, there were 90 total blank samples.

```python
blank_train_ids = []
blank_test_ids = []
for i in range(X_train.shape[0]):
  if np.sum(X_train[i,:,:,0])==0 or np.sum(X_train[i,:,:,1])==0 or np.sum(X_train[i,:,:,2])==0.:
    blank_train_ids.append(i)
  if i < X_test.shape[0] and (np.sum(X_test[i,:,:,0])==0 or np.sum(X_test[i,:,:,1])==0 or np.sum(X_test[i,:,:,2])==0.):
    blank_test_ids.append(i)
X_train = np.delete(X_train, blank_train_ids, 0)
y_train = np.delete(y_train, blank_train_ids, 0)
X_test = np.delete(X_test, blank_test_ids, 0)
y_test = np.delete(y_test, blank_test_ids, 0)

print(f'Post blank removal ({len(blank_train_ids) + len(blank_test_ids)} total removed)')
print(f'Training: {len(X_train)} samples ({sum(y_train)} "Real" | {len(X_train)-sum(y_train)} "Bogus") \n Testing: {len(X_test)} samples ({sum(y_test)} "Real" | {len(X_test)-sum(y_test)} "Bogus")')
```

## Construct Model
Construct a model architecture to train

```python
keras.backend.clear_session()  # Clear any previous sessions
model = Sequential([  # Construct a sequential Convolutional Neural Network
		Conv2D(64, (3,3),
			input_shape=(31,31,3)),
		MaxPooling2D(pool_size=(4,4)),
		GlobalAveragePooling2D(),
		Dense(128, activation='relu'),
		Dense(64, activation='relu'),
		Dense(1, activation='sigmoid')
], name='Demo_Model')
model.summary()
```

## Define MCC and F1 Score functions

```python
def MCC(y_true, y_pred):
    y_pred_pos = K.round(K.clip(y_pred, 0, 1))
    y_pred_neg = 1 - y_pred_pos

    y_pos = K.round(K.clip(y_true, 0, 1))
    y_neg = 1 - y_pos

    tp = K.sum(y_pos * y_pred_pos)
    tn = K.sum(y_neg * y_pred_neg)

    fp = K.sum(y_neg * y_pred_pos)
    fn = K.sum(y_pos * y_pred_neg)

    numerator = (tp * tn - fp * fn)
    denominator = K.sqrt((tp + fp) * (tp + fn) * (tn + fp) * (tn + fn))

    return numerator / (denominator + K.epsilon())
    
    
def f1_score(y_true, y_pred): #taken from old keras source code
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val
```

## Compile Model

```python
### TRAINING PARAMETERS ###
epochs = 189  # Number of training epochs
bs = 50  # Batch size
learning_rate = 1e-4  # Learning rate gamma
opt = 'adam'  # Adam optimizer
loss = 'binary_crossentropy'  # Loss function
metrics = ['accuracy', f1_score, MCC]  # Performance metrics
val_split = 0.1  # Validation split
stopping_monitor = 'val_loss'  # Early stopping will monitor validation loss
stopping_patience = 30  # Number of epochs early stopping will consider
es = EarlyStopping(monitor=stopping_monitor, mode='min', verbose=1, patience=stopping_patience)

### COMPILE MODEL ###
model.compile(optimizer=opt,  # Compile model with optimizer, loss function and performance metrics
		loss=loss,
		metrics=metrics)

history = model.fit(x=X_train,  # Fit the model with training data with batch size, epochs and validation split. Will stop early if early stopping criteria is met
			y=y_train,
			batch_size=bs,
			epochs=epochs,
			validation_split=val_split,
			callbacks=[es])

### SAVE MODEL ###
model.save("./Models/demo_model.h5")  # Save the currently trained model & weights
model.save_weights("./Models/demo_model_weights.h5")
json.dump(history.history, open('./Models/training_history.json', 'w'))
```

## Training History

### Plotting Functions

```python
def PlotAccuracyOverEpoch(history_dict, figname='Model Accuracy over Epoch'):
  plt.figure(figname, figsize=(7,4), dpi=120)
  plt.plot(history_dict["accuracy"], label='Training', color='tab:blue', lw=1)
  plt.plot(history_dict["val_accuracy"], label='Validation', color='tab:orange', lw=1)
  plt.grid()
  plt.xlabel('Epoch')
  plt.ylabel('Accuracy')
  plt.title(figname)
  plt.legend()
  plt.show()

def PlotLossOverEpoch(history_dict, plotnow=True, figname='Model Loss over Epoch'):
  plt.figure(figname, figsize=(7,4), dpi=120)
  plt.plot(history_dict["loss"], label='Training', color='tab:blue', lw=1)
  plt.plot(history_dict["val_loss"], label='Validation', color='tab:orange', lw=1)
  plt.grid()
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.title(figname)
  plt.legend()
  plt.show()

def PlotF1MCCOverEpoch(history_dict, plotnow=True): 
  fig, ax1 = plt.subplots(figsize=(7,4), dpi=120)
  ax1.plot(history_dict["f1_score"], label='Training', color='tab:blue', lw=1)
  ax1.plot(history_dict["val_f1_score"], label='Validation', color='tab:blue', lw=1, ls='--')
  ax1.set_xlabel('Epoch')
  ax1.set_ylabel('F1 Score', color='tab:blue')
  ax1.tick_params(axis='y', labelcolor='tab:blue')
  ax2 = ax1.twinx()
  ax2.plot(history_dict["MCC"], color='tab:orange', lw=1)
  ax2.plot(history_dict["val_MCC"], color='tab:orange', lw=1, ls='--')  
  ax2.set_ylabel('MCC', color='tab:orange')
  ax2.tick_params(axis='y', labelcolor='tab:orange')
  plt.title("MCC & F1 Score over Epochs")
  #fig.tight_layout()
  ax1.legend()
  plt.show()

def PlotROC(fpr, tpr, auc_model, plotnow=True, figname='ROC Curve', plotinset=False, inset=0.1):
  plt.figure(figname, figsize=(7,7), dpi=120)
  plt.plot([0, 1], [0, 1], 'k--')
  plt.plot(fpr, tpr, label='ROC (area = {:.3f})'.format(auc_model), color='tab:blue')
  plt.xlabel('False positive rate (Contamination)')
  plt.ylabel('True positive rate (Sensitivity)')
  plt.title('ROC curve')
  plt.legend(loc='best')
  plt.grid()
  plt.show()

def PlotMisclassificationOverThreshold(test_labels, predictions, plotnow=True, figname='Misclassifications Over Thresholds'):
  ### Credit: Dmitry Duev's braai algorithm: https://github.com/dmitryduev/braai/blob/master/nb/braai_train.ipynb
  rbbins = np.arange(-0.0001, 1.0001, 0.0001)
  bogus_scores = []
  real_scores = []
  for i in range(len(test_labels)):
    if test_labels[i]==1:
      real_scores.append(predictions[i][0])
    else:
      bogus_scores.append(predictions[i][0])
  h_b, e_b = np.histogram(bogus_scores, bins=rbbins, density=True)
  h_b_c = np.cumsum(h_b)
  h_r, e_r = np.histogram(real_scores, bins=rbbins, density=True)
  h_r_c = np.cumsum(h_r)

  fig = plt.figure(figname, figsize=(7, 4), dpi=120)
  ax = fig.add_subplot(111)

  rb_thres = np.array(list(range(len(h_b)))) / len(h_b)

  ax.plot(rb_thres, h_r_c/np.max(h_r_c), 
          label='False Negative Rate (FNR)', linewidth=1.5)
  ax.plot(rb_thres, 1 - h_b_c/np.max(h_b_c), 
          label='False Positive Rate (FPR)', linewidth=1.5)

  mmce = (h_r_c/np.max(h_r_c) + 1 - h_b_c/np.max(h_b_c))/2
  ax.plot(rb_thres, mmce, '--',
          label='Mean misclassification error', color='gray', linewidth=1.5)

  ax.set_xlim([-0.05, 1.05])

  ax.set_xticks(np.arange(0, 1.1, 0.1))
  ax.set_yticks(np.arange(0, 1.1, 0.1))

  ax.set_yscale('log')
  ax.set_ylim([5e-4, 1])
  vals = ax.get_yticks()
  ax.set_yticklabels(['{:,.1%}'.format(x) if x < 0.01 else '{:,.0%}'.format(x) for x in vals])

  # thresholds:
  min_mmce_thrs = np.where(mmce==np.min(mmce))[0][0]
  fpr_fnr_diff = abs((h_r_c/np.max(h_r_c))-(1 - h_b_c/np.max(h_b_c)))
  min_diff = np.min(fpr_fnr_diff)
  mmce_thrs_cross = np.where(fpr_fnr_diff==min_diff)[0][0]
  thrs = [0.5,rb_thres[mmce_thrs_cross]]
  print('Threshold | FNR (%) | FPR (%) | MMCE (%)')
  for t in thrs:
      m_t = rb_thres < t
      fnr = np.array(h_r_c/np.max(h_r_c))[m_t][-1]
      fpr = np.array(1 - h_b_c/np.max(h_b_c))[m_t][-1]
      print('{:,.2} | {:,.1%} | {:,.1%} | {:,.1%}'.format(t, fnr, fpr, (fnr + fpr)/2))
      # ax.vlines(t_1, 0, 1.1)
      ax.vlines(t, 0, max(fnr, fpr))
      ax.text(t - .05, max(fnr, fpr) + 0.01, f' {fnr*100:.1f}% FNR\n {fpr*100:.1f}% FPR', fontsize=10)

  ax.set_xlabel('RB score threshold')
  ax.set_ylabel('Cumulative percentage')
  ax.legend(loc='lower center')
  ax.grid(True, which='major', linewidth=.5)
  ax.grid(True, which='minor', linewidth=.3)
  plt.tight_layout()
  plt.title(figname)
  plt.show()
  
  return rb_thres[mmce_thrs_cross]

def PlotPredictionHistogram(test_labels, predictions, plotnow=True, figname='Prediction Score Historgram', bins=50):
  bogus_scores = []
  real_scores = []
  for i in range(len(test_labels)):
    if test_labels[i]==1:
      real_scores.append(predictions[i][0])
    else:
      bogus_scores.append(predictions[i][0])

  fig = plt.figure(figname, figsize=(7,4), dpi=120)
  ax = fig.add_subplot(111)
  ax.grid()
  ax.set_axisbelow(True)
  ax.hist(real_scores, bins=bins, color='green', histtype=u'step', lw=1, label='Real')
  ax.hist(bogus_scores, bins=bins, color='red', histtype=u'step', lw=1, label='Bogus')
  plt.xlabel('RB Score')
  plt.ylabel('Count')
  plt.xlim([-0.05, 1.05])
  plt.legend()
  plt.show()

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues,
                          plotnow=True):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
        plt.figure('Confusion Matrix')
        normal = matplotlib.colors.Normalize(vmin=0, vmax=1)
        plt.imshow(cm, interpolation='nearest', cmap=cmap, norm=normal)
    else:
        print('Confusion matrix, without normalization')
        plt.figure('Confusion Matrix')
        plt.imshow(cm, interpolation='nearest', cmap=cmap)

    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 1.5
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.grid(False)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

## Function Definitions

```python
def PlotTriplet(a, data, labels=["none"], figname='Triplet'):
  fig, (ax1, ax2, ax3) = plt.subplots(1,3)
  if labels != ["none"]:
    ax2.set_title("{} (label: {})".format(figname,labels[a]))
  else:
    ax2.set_title(figname)
  ax1.imshow(data[a,:,:,2], cmap='gray', vmin=0, vmax=1)
  ax2.imshow(data[a,:,:,0], cmap='gray', vmin=0, vmax=1)
  ax3.imshow(data[a,:,:,1], cmap='gray', vmin=0, vmax=1)
  print("_________________________________________________")
  plt.show()

def PlotSamples(test_data, test_labels, predictions, num_samples=0, greaterthan=0.8, lessthan=0.1):
  counter = 0
  for l in range(len(test_labels)):
    if counter >= num_samples:
      break
    if greaterthan > lessthan:
      if predictions[l] > greaterthan or predictions[l] < lessthan:
       PlotTriplet(l, test_data, figname='Label: {} | Score: {:,.2}'.format(test_labels[l], predictions[l][0]))
       counter += 1
    elif predictions[l] > greaterthan and predictions[l] < lessthan:
      PlotTriplet(l, test_data, figname='Label: {} | Score: {:,.2}'.format(test_labels[l], predictions[l][0]))
      counter += 1
```
