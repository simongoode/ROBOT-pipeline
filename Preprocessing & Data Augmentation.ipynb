## ConvertLabels Function

Convert label markers (e.g. "S", "B", "F") into label values (e.g. 0, 1)

```python
def ConvertLabels(a, positive_label_val, negative_label_val, positive_markers, negative_markers):
  arr = []
  for l in range(len(a)):
    if a[l] in positive_markers:
      arr.append(positive_label_val)
    elif a[l] in negative_markers:
      arr.append(negative_label_val)
    else:
      raise ValueError("Label not recognised: {}".format(a[l]))
  return arr
```

## Normalize data

```python
def PercentileNormalize(a, percentile=100, minimum=0, a_name='array'):
  print("Percentile ({}) of {}: {}".format(percentile, a_name, np.percentile(a, percentile)))
  print("Minimum: {}, Maximum: {}".format(np.min(a), np.max(a)))
  return (a-minimum)/(np.percentile(a, percentile)-minimum)
```

```python
def PlotTriplet(a, data, labels=None, figname='Triplet', plotnow=True):
  fig, (ax1, ax2, ax3) = plt.subplots(1,3)
  if labels.any != None:
    fig.suptitle("{} (label: {})".format(figname,labels[a]))
  else:
    fig.suptitle(figname)
  ax1.set_title('Template')
  ax1.imshow(data[a,:,:,2], cmap='gray', vmin=0, vmax=1)
  ax2.set_title('Science')
  ax2.imshow(data[a,:,:,0], cmap='gray', vmin=0, vmax=1)
  ax3.set_title('Subtraction')
  ax3.imshow(data[a,:,:,1], cmap='gray', vmin=0, vmax=1)
  if plotnow==True:
    plt.show()
```

```python
pct = 98.3945  # Value where sci pctle is approx temp pctle
sci_norm = PercentileNormalize(data_RAW[:,:,:,0],
                               percentile=pct,
                               #minimum=np.min(data_RAW[:,:,:,0]),
                               a_name='science data')
print("---: {}, ---: {}\n".format(np.min(sci_norm), np.max(sci_norm)))
sub_norm = PercentileNormalize(data_RAW[:,:,:,1],
                               percentile=pct,
                               #minimum=np.min(data_RAW[:,:,:,1]),
                               a_name='subtraction data')
print("---: {}, ---: {}\n".format(np.min(sub_norm), np.max(sub_norm)))
temp_norm = PercentileNormalize(data_RAW[:,:,:,2],
                                percentile=pct,
                               #minimum=np.min(data_RAW[:,:,:,2]),
                               a_name='template data')
print("---: {}, ---: {}\n".format(np.min(temp_norm), np.max(temp_norm)))

data_NORMED = np.ndarray(data_RAW.shape)
data_NORMED[:,:,:,0] = sci_norm
data_NORMED[:,:,:,1] = sub_norm
data_NORMED[:,:,:,2] = temp_norm
print(data_NORMED.shape)
del(sci_norm, sub_norm, temp_norm)
```

## Data Generators

Separate real samples from bogus samples

### CropCenter function

```python
def CropCenter(image, crop_width, crop_height):
	#print(image.shape)
	centroid = (int(image.shape[0]/2), int(image.shape[1]/2))
	#print(centroid)
	
	if (int(crop_width)/2).is_integer():
		crop_xbounds = (int(centroid[0]-crop_width/2), int(centroid[0]+crop_width/2))
	else:
		crop_xbounds = (int(centroid[0]-crop_width/2+0.5), int(centroid[0]+crop_width/2+0.5))
	if crop_xbounds[0] < 0 or crop_xbounds[1] > image.shape[0]:
		raise ValueError("Crop dimensions exceed the boundaries of original image")
	
	if (int(crop_height)/2).is_integer():
		crop_ybounds = (int(centroid[1]-crop_height/2), int(centroid[1]+crop_height/2))
	else:
		crop_ybounds = (int(centroid[1]-crop_height/2+0.5), int(centroid[1]+crop_height/2+0.5))
	if crop_ybounds[0] < 0 or crop_ybounds[1] > image.shape[1]:
		raise ValueError("Crop dimensions exceed the boundaries of original image")
			
	return image[crop_xbounds[0]:crop_xbounds[1], crop_ybounds[0]:crop_ybounds[1]]
```

### Augmentation parameters

```python
samples_to_generate = 5000
batch_size = 500
epochs = samples_to_generate//batch_size
crop_dims = (31,31)

real_data = data_NORMED[df.loc[df['Converted Labels'] == 1, 'Sample IDs']]
bogus_data = data_NORMED[df.loc[df['Converted Labels'] == 0, 'Sample IDs']]
print("Reals: {}".format(real_data.shape))
print("Boguses: {}".format(bogus_data.shape))
```

### Real Generator

```python
real_data_generator = ImageDataGenerator(rotation_range=360,
                                         width_shift_range=3,
                                         height_shift_range=3,
                                         horizontal_flip=True,
                                         vertical_flip=True)

real_aug_data = np.empty((samples_to_generate, 121, 121, 3))
for e in range(epochs):
  x = e*batch_size
  y = (e+1)*batch_size
  real_datagen_iterator = real_data_generator.flow(real_data,
                                                 np.ones(real_data.shape[0], dtype=int),
                                                 batch_size=batch_size)
  real_aug_data[x:y] = (real_datagen_iterator.next()[0])

real_aug_cropped_data = np.empty((real_aug_data.shape[0],
                                   crop_dims[0],
                                   crop_dims[1],
                                   3))
for sample in range(real_aug_data.shape[0]):
  for channel in range(3):
    real_aug_cropped_data[sample,:,:,channel] = CropCenter(real_aug_data[sample,:,:,channel], crop_dims[0], crop_dims[1])
del(real_aug_data)
print(real_aug_cropped_data.shape)
```

### Bogus Generator

```python
bogus_data_generator = ImageDataGenerator(width_shift_range=3,
                                         height_shift_range=3,
                                         horizontal_flip=True,
                                         vertical_flip=True)

bogus_datagen_iterator = bogus_data_generator.flow(bogus_data,
                                                 np.zeros(bogus_data.shape[0], dtype=int),
                                                 batch_size=batch_size)

bogus_aug_data = np.empty((samples_to_generate, 121, 121, 3))
for e in range(epochs):
  x = e*batch_size
  y = (e+1)*batch_size
  bogus_datagen_iterator = real_data_generator.flow(bogus_data,
                                                 np.ones(bogus_data.shape[0], dtype=int),
                                                 batch_size=batch_size)
  bogus_aug_data[x:y] = (bogus_datagen_iterator.next()[0])

bogus_aug_cropped_data = np.empty((bogus_aug_data.shape[0],
                                   crop_dims[0],
                                   crop_dims[1],
                                   3))
for sample in range(bogus_aug_data.shape[0]):
  for channel in range(3):
    bogus_aug_cropped_data[sample,:,:,channel] = CropCenter(bogus_aug_data[sample,:,:,channel], crop_dims[0], crop_dims[1])
del(bogus_aug_data)
print(bogus_aug_cropped_data.shape)
```

### Merge generated data & labels to create dataset 2

```python
data_concat = np.concatenate((real_aug_cropped_data, bogus_aug_cropped_data))
labels_concat = np.concatenate((np.ones(samples_to_generate, dtype=int), np.zeros(samples_to_generate, dtype=int)))
dataset = list(zip(data_concat, labels_concat))
random.shuffle(dataset)
```
